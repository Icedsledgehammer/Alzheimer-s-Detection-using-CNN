This is where you setup the TensorFlow to work with Google Cloud's TPU hardware for machine learning tasks

```python
try:
    tpu = tf.distribute.cluster_resolver.TPUClusterResolver()
    print("Device:", tpu.master())
    tf.config.experimental_connect_to_cluster(tpu)
    tf.tpu.experimental.initialize_tpu_system(tpu)
    strategy = tf.distribute.experimental.TPUStrategy(tpu)
    
except:
    strategy = tf.distribute.get_strategy()
    
print('Number of replicas:', strategy.num_replicas_in_sync)

print(tf.version)
```

Set constant variables instead of hard coding numbers into the code. Saves time when certain parameters have to be changed 

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE
BATCH_SIZE = 16 * strategy.num_replicas_in_sync
IMAGE_SIZE = [176, 208]
EPOCHS = 30
```

Load the dataset for training, here 90% is training and 10% is validation 

```python
#Load the dataset file to train it
train_ds = tf.keras.preprocessing.image_dataset_from_directory("C:\\Python Workspace\\Image processing\\Alzheimer's dataset\\archive (3)\\Alzheimer_s Dataset\\train",
                                                               validation_split = 0.1,
                                                               subset = "training",
                                                               seed = 1337,
                                                               image_size = IMAGE_SIZE,
                                                               batch_size = BATCH_SIZE)

val_ds = tf.keras.preprocessing.image_dataset_from_directory("C:\\Python Workspace\\Image processing\\Alzheimer's dataset\\archive (3)\\Alzheimer_s Dataset\\train",
                                                             validation_split = 0.1,
                                                             subset = "validation",
                                                             seed = 1337,
                                                             image_size = IMAGE_SIZE,
                                                             batch_size = BATCH_SIZE)
```

Name all the classes in which you'll be doing your magic. Here we have 4

```python
class_names = ['MildDementia', 'ModerateDementia', 'NonDementia', 'VeryMildDementia']
train_ds.class_names = class_names
val_ds.class_names = class_names

NUM_CLASSES = len(class_names)
```

Visualization of data 

```python
plt.figure(figsize = (8, 8))
for image, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(image[i].numpy().astype("uint8"))
        plt.title(train_ds.class_names[labels[i]])
        plt.axis("off")
```

Feature engineering, this is where we convert our model into one-hot encodings. One-hot encodings are a way for the model to understand that we're looking at categorial instead of continuous data.

```python
def one_hot_label(image, label):
    label = tf.one_hot(label, NUM_CLASSES)
    return image, label 

train_ds = train_ds.map(one_hot_label, num_parallel_calls = AUTOTUNE)
val_ds = val_ds.map(one_hot_label, num_parallel_calls = AUTOTUNE)
```

The below cell makes the calling of images more efficient

```python
train_ds = train_ds.cache().prefetch(buffer_size = AUTOTUNE)
val_ds = train_ds.cache().prefetch(buffer_size = AUTOTUNE)
```

```python
NUM_IMAGES = []

for label in class_names:
    # Construct the directory path with a different approach
    dir_name = os.path.join("C:\\Python Workspace\\Image processing\\Alzheimer's dataset\\archive (3)\\Alzheimer_s Dataset\\train", label[:-2] + 'ed')
    
    # Check if the directory exists
    if os.path.exists(dir_name):
        num_files = len(os.listdir(dir_name))
        NUM_IMAGES.append(num_files)
    else:
        print(f"Directory not found: {dir_name}")

# Check the NUM_IMAGES list to see the counts
print(NUM_IMAGES)
```

Now to build a model, here we'll be using CNN(Convulational Neural Network)

```python
def conv_block(filters):
    block = tf.keras.Sequential([
           tf.keras.layers.SeparableConv2D(filters, 3, activation = 'relu', padding = 'same'),
           tf.keras.layers.SeparableConv2D(filters, 3, activation = 'relu', padding = 'same'),
           tf.keras.layers.MaxPool2D()])
    
    return block
```

```python
def dense_block(units, dropout_rate):
    block = tf.keras.Sequential([
           tf.keras.layers.Dense(units, activation = 'relu'),
           tf.keras.layers.BatchNormalization(),
           tf.keras.layers.Dropout(dropout_rate)])
    
    return block
```

```python
def build_model():
    model = tf.keras.Sequential(
        [
            tf.keras.Input(shape = (* IMAGE_SIZE, 3)),
            
            tf.keras.layers.Conv2D(16, 3, activation = "relu", padding = "same"),
            tf.keras.layers.Conv2D(16, 3, activation = "relu", padding = "same"),
            tf.keras.layers.MaxPool2D(),
            
            conv_block(32),
            conv_block(64),
            
            conv_block(128),
            tf.keras.layers.Dropout(0.1),
            
            conv_block(256),
            tf.keras.layers.Dropout(0.1),
            
            tf.keras.layers.Flatten(),
            dense_block(512, 0.7),
            dense_block(128, 0.5),
            dense_block(64, 0.3),
            
            tf.keras.layers.Dense(NUM_CLASSES, activation = "softmax")
        ]
    )
    
    return model
```

```python
with strategy.scope():
    model = build_model()
    
    METRICS = [tf.keras.metrics.AUC(name = 'auc')]
    
    model.compile(
        optimizer = 'adam',
        loss = tf.losses.CategoricalCrossentropy(),
        metrics = METRICS
    )
```

Now to train the model. 

To more efficiently train our model. We will be using callbacks to adjust our learning rate and to stop our model once it converges.

The learning rate is a very important hyperparameter in the model. Having a LR that is too high will prevent the model from converging. Having a LR that is too slow will make the process too long. Stopping our model early is one mechanism that prevents overfitting.

```python
def exponential_decay(lr0, s):
    def exponential_decay_fn(epoch):
        return lr0 * 0.1 **(epoch / s)
    return exponential_decay_fn

exponential_decay_fn = exponential_decay(0.01, 20)

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(exponential_decay_fn)

checkpoint_cb = tf.keras.callbacks.ModelCheckpoint("alzheimer_model.h5",
                                                    save_best_only=True)

early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience = 10,
                                                     restore_best_weights = True)
```

Now to fit our model

```python
history = model.fit(
    train_ds,
    validation_data = val_ds,
    callbacks = [checkpoint_cb, early_stopping_cb, lr_scheduler],
    epochs = EPOCHS
)
```

Visualization of models

```python
fig, ax = plt.subplots(1, 2, figsize = (20, 3))
ax = ax.ravel()

for i, met in enumerate(['auc', 'loss']):
    ax[i].plot(history.history[met])
    ax[i].plot(history.history['val_' + met])
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['Train', 'Val'])
```
